```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})

```

```python
df = pd.read_csv('datalimpia.csv')
```

```python
df.head()
```

```python
df1 = pd.concat([df['Page Name'].value_counts(), 
                df['Page Name'].value_counts(normalize=True).mul(100)],axis=1, keys=('Cantidad','Porcentaje'))
print (df1)
```

```python
sns.scatterplot(x = 'Page Likes at Posting', y = 'Page Name', data = df)
plt.xlabel('Likes fanpage')
plt.ylabel('Medios de comunicación')
plt.title('Likes / Medios de comunicación');
```

```python
df['Type'].value_counts(normalize = True)
```

```python
sns.countplot(df['Type']); 
plt.xlabel('Tipos de formatos utilizados')
plt.ylabel('Cantidad de personas')
plt.title('Tipos de publicación en Facebook')
```

```python
sns.scatterplot(x = 'Likes', y = 'Page Name', data = df)
plt.xlabel('Likes fanpage')
plt.ylabel('Medios de comunicación')
plt.title('Likes / Medios de comunicación');
```

```python
#Likes 

print(np.mean(df['Likes']))
print(np.max(df['Likes']))
print(np.min(df['Likes']))
print(np.median(df['Likes']))
```

```python
sns.scatterplot(x = 'Comments', y = 'Page Name', data = df)
plt.xlabel('Likes fanpage')
plt.ylabel('Medios de comunicación')
plt.title('Likes / Medios de comunicación');
```

```python
sns.scatterplot(x = 'Shares', y = 'Page Name', hue = 'Overperforming Score', data = df, s = 100)
plt.xlabel('Likes fanpage')
plt.ylabel('Medios de comunicación')
plt.title('Likes / Medios de comunicación');
```
